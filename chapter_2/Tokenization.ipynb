## Regex Tokenization

#### English

```python
import nltk
```

```python
text = "That         [U.S.A.] poster-print costs $12.40... #elections"

pattern = r''' (?x)    # verbose regexps flag to tell the compiler ignore comments and white spaces
  (?:[A-Z]\.)+         # abbreviations, e.g. U.S.A.
| \w+(?:-\w+)*         # words with optional internal hyphens
| \$?\d+(?:\.\d+)?%?   # currency, percentage, e.g. $34, 34.3%
| \.{3}                # ellipsis
| [][.,;"#'?():_'-]    # Some seperate tokens (note ][ is included)
'''

nltk.regexp_tokenize(text, pattern)
```

#### Arabic

```python
text = "محمدٌ سيد     #الكونين [والثقلين]، بنور محمد-النَّبي الأمي- $12.51...."

pattern = r''' (?x)    # verbose regexps flag to tell the compiler ignore comments and white spaces
  (?:[A-Z]\.)+         # abbreviations, e.g. U.S.A.
| \w+(?:-\w+)*         # words with optional internal hyphens
| \$?\d+(?:\.\d+)?%?   # currency, percentage, e.g. $34, 34.3%
| \.{3}                # ellipsis
| [][.,;"#'?():_'-]    # Some seperate tokens (note ][ is included)
'''

nltk.regexp_tokenize(text, pattern)
```

#### Essential Functions

```python
def get_stats(ids):
    # Count the occurences of consecutive pairs
    counts = {}
    for pair in zip(ids, ids[1:]): # Pythonic way to iterate consecutive elements
        counts[pair] = counts.get(pair, 0) + 1
    return counts

stats = get_stats(tokens_ids)
print(sorted(((v, k) for k, v in stats.items()), reverse=True))
top_pair = max(stats, key=stats.get)
chr(top_pair[0]), chr(top_pair[1]) # Print the characters with the most consecutive occurences
```

```python
def merge(ids, pair, index):
    # Merge consecutive pairs and give them a new index
    newids = []
    i = 0
    while i < len(ids):
        if i < len(ids) -1 and ids[i] == pair[0] and ids[i+1] == pair[1]:
            newids.append(index)
            i += 2
        else:
            newids.append(ids[i])
            i += 1
    return newids

tokens_ids = merge(tokens_ids, top_pair, 256)
print(tokens_ids)
print("Length =", len(tokens_ids))
```

#### Training the tokenizer parameters (merges)

```python
vocabsize = 276 # Hyperparameter
num_merges = vocabsize - 256 # 20 merges
cp_ids = list(tokens_ids) # copy of tokens_ids

merges = {} # Tokenizer Parameters
for i in range(num_merges):
    # find the most occurring pair
    stats = get_stats(cp_ids)
    top_pair = max(stats, key=stats.get)
    index = 256 + i
    print(f"merging {top_pair} into {index}")
    cp_ids = merge(cp_ids, top_pair, index)
    merges[top_pair] = index
```

```python
print("Before merging:", len(tokens_ids))
print("After merging:", len(cp_ids))
print(f"Compression Ratio: {len(tokens_ids)/len(cp_ids):0.2f}X")
```

#### Decoding Text

```python
vocab = {idx: bytes([idx]) for idx in range(256)}
print(len(vocab))
for (p0,p1), idx in merges.items():
    # Add the new tokens to the vocab
    vocab[idx] = vocab[p0] + vocab[p1] # This is just a concatenation of two bytes objects
print(len(vocab))
```

```python
def decode(ids):
    tokens = b"".join(vocab[idx] for idx in ids) # concatenate bytes
    text = tokens.decode("utf-8", errors="replace") # error=replace to replace undefined unicodes with "?"
    return text
# print(decode([128])) # to check the errors="replace"
print(decode(tokens_ids))
```

#### Encoding


```python
def encode(text):
    tokens_ids = list(text.encode("utf-8"))
    while len(tokens_ids) >= 2:
        stats = get_stats(tokens_ids)
        pair = min(stats, key=lambda p: merges.get(p, float("inf"))) # from stats get the pair that have the minimum index in merges
        if pair not in merges:
            break # noting else can be merged
        idx = merges[pair]
        tokens_ids = merge(tokens_ids, pair, idx)
    return tokens_ids

print(encode(" "))
print(decode(encode("محمد صلى الله عليه وسلم")))
text_backward = decode(encode(text))
print(text_backward)
print(text_backward == text)
```
